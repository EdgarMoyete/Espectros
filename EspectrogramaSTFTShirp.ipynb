#### Espectrograma sin y con STFT de shirp

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.signal import chirp, spectrogram
```

```python
fs = 250
time = np.arange(0, 10, 1/fs)
sig = chirp(time, f0=6, f1=1, t1=10)
sig.shape
```

```python
plt.plot(time, sig)
plt.grid()
plt.xlabel('Time')
plt.ylabel('Amplitude')
plt.show()
```

```python
powerSpectrum, freqenciesFound, t, imageAxis = plt.specgram(sig, Fs=fs)
plt.xlabel('Time [sec]')
plt.ylabel('Frequency [Hz]')
plt.show()
```

```python
plt.pcolormesh(t, freqenciesFound, powerSpectrum)
plt.xlabel('Time [sec]')
plt.ylabel('Frequency [Hz]')
plt.show()
```

```python
f, t, Sxx = spectrogram(sig, fs, window='hann', nperseg=250, noverlap=225)
plt.pcolormesh(t, f, Sxx)
plt.xlabel('Time [sec]')
plt.ylabel('Frequency [Hz]')
plt.show()
```
