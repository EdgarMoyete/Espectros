#### CWT

```python
import numpy as np
import matplotlib.pyplot as plt
import pywt
```

```python
#8 segundos de una señal con fs=250Hz
#la señal es la suma de tres señales senos

fs = 250
time = np.arange(0, 8, 1/fs)
x1 = np.sin(2*np.pi*4*time) #seno a 4Hz
x2 = np.sin(2*np.pi*16*time) #seno a 16Hz
x3 = np.sin(2*np.pi*32*time) #seno a 32Hz
sig = x1 + x2 + x3
```

```python
plt.plot(time, sig)
plt.grid()
plt.xlabel('Time')
plt.ylabel('Amplitude')
plt.show()
```

#### Morlet

```python
frecuencias = pywt.scale2frequency('morl', [2,4,8,16,32,64,128,256]) / (1/fs)
print("Frecuencias para diferentes escalas morl: \n", frecuencias)
```

```python
coef, freqs = pywt.cwt(sig, np.arange(1,129),'morl')
```

```python
print(coef.shape)
plt.matshow(coef)
plt.show()
```

```python
plt.pcolormesh(time, freqs, coef)
plt.xlabel('Time [sec]')
plt.ylabel('Frequency [Hz]')
plt.show()
```

```python
plt.pcolormesh(time, freqs, coef)
plt.axis([0, time[-1:], freqs[-1:], 0.2])
plt.xlabel('Time [sec]')
plt.ylabel('Frequency [Hz]')
plt.show()
```

```python
print("CWT a escala de grises, esto es lo que se va a clasificar")
plt.imshow(coef, cmap=plt.cm.binary)
plt.show()
```

```python
print(freqs.shape)
print(freqs)
```

#### Complex Morlet

```python
frecuencias = pywt.scale2frequency('cmor3-3', [2,4,8,16,32,64,128,256]) / (1/fs)
print("Frecuencias para diferentes escalas cmor3-3: \n", frecuencias)
```

```python
coef, freqs = pywt.cwt(sig, np.arange(1,257),'cmor3-3')
```

```python
print(coef.shape)
plt.matshow(abs(coef))
plt.show()
```

```python
plt.pcolormesh(time, freqs, abs(coef))
plt.xlabel('Time [sec]')
plt.ylabel('Frequency [Hz]')
plt.show()
```

```python
plt.pcolormesh(time, freqs, abs(coef))
plt.axis([0, time[-1:], freqs[-1:], 0.16])
plt.xlabel('Time [sec]')
plt.ylabel('Frequency [Hz]')
plt.show()
```

```python
print("CWT a escala de grises, esto es lo que se va a clasificar")
plt.imshow(abs(coef), cmap=plt.cm.binary)
plt.show()
```
