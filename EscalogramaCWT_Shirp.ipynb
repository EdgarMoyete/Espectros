#### CWT de shirp

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.signal import chirp
import pywt
```

```python
fs = 250
time = np.arange(0, 10, 1/fs)
sig = chirp(time, f0=6, f1=1, t1=10)
sig.shape
```

```python
plt.plot(time, sig)
plt.grid()
plt.xlabel('Time')
plt.ylabel('Amplitude')
plt.show()
```

#### Morlet

```python
frecuencias = pywt.scale2frequency('morl', [2,4,8,16,32,64,128,256]) / (1/fs)
print("Frecuencias para diferentes escalas morl: \n", frecuencias)
```

```python
coef, freqs = pywt.cwt(sig, np.arange(1,513),'morl')
```

```python
print(coef.shape)
plt.matshow(coef)
plt.show()
```

```python
plt.pcolormesh(time, freqs, coef)
plt.xlabel('Time [sec]')
plt.ylabel('Frequency [Hz]')
plt.show()
```

```python
plt.pcolormesh(time, freqs, coef)
plt.axis([0, time[-1:], freqs[-1:], 0.04])
plt.xlabel('Time [sec]')
plt.ylabel('Frequency [Hz]')
plt.show()
```

```python
print("CWT a escala de grises, esto es lo que se va a clasificar")
plt.imshow(coef, cmap=plt.cm.binary)
plt.show()
```

```python
print(freqs.shape)
print(freqs)
```

#### Complex Morlet

```python
frecuencias = pywt.scale2frequency('cmor3-3', [2,4,8,16,32,64,128,256]) / (1/fs)
print("Frecuencias para diferentes escalas cmor3-3: \n", frecuencias)
```

```python
coef, freqs = pywt.cwt(sig, np.arange(1,513), 'cmor3-3')
```

```python
print(coef.shape)
plt.matshow(abs(coef))
plt.show()
```

```python
plt.pcolormesh(time, freqs, abs(coef))
plt.xlabel('Time [sec]')
plt.ylabel('Frequency [Hz]')
plt.show()
```

```python
plt.pcolormesh(time, freqs, abs(coef))
plt.axis([0, time[-1:], freqs[-1:], 0.04])
plt.xlabel('Time [sec]')
plt.ylabel('Frequency [Hz]')
plt.show()
```
