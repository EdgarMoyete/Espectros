#### Escalograma con CWT de seno

```python
import numpy as np
import matplotlib.pyplot as plt
import pywt
```

```python
# 8 segundos de una señal con fs=250Hz
# la señal es la suma de tres señales senos

fs = 250
time = np.arange(0, 8, 1/fs)
x1 = np.sin(2*np.pi*4*time) #seno a 4Hz
x2 = np.sin(2*np.pi*16*time) #seno a 16Hz
x3 = np.sin(2*np.pi*32*time) #seno a 32Hz
sig = x1 + x2 + x3
```

```python
plt.plot(time, sig)
plt.grid()
plt.xlabel('Time')
plt.ylabel('Amplitude')
plt.show()
```

### Morlet

```python
# calcular las escalas de 2-34Hz para Morlet wavelets
# f = scale2frequency(wavelet, scale)/sampling_period
escalas = pywt.scale2frequency('morl', np.arange(2,35)) / (1/fs)
print(escalas)
```

```python
# Escalograma (CWT)
# pywt.cwt(senal, escalas, tipo wavelet, periodo de muestreo = (1 / fs))
coef, freqs = pywt.cwt(sig, escalas, 'morl', sampling_period = (1 / fs))
```

```python
plt.pcolormesh(time, freqs, coef)
plt.xlabel('Time [sec]')
plt.ylabel('Frequency [Hz]')
plt.show()
```

```python
print(coef.shape)
plt.matshow(coef)
plt.show()
```

### Complex Morlet

```python
# calcular las escalas de 2-34Hz para complex Morlet wavelets
# f = scale2frequency(wavelet, scale)/sampling_period
escalas = pywt.scale2frequency('cmor3-3', np.arange(2,35)) / (1/fs)
print(escalas)
```

```python
# Escalograma (CWT)
# pywt.cwt(senal, escalas, tipo wavelet, periodo de muestreo = (1 / fs))
coef, freqs = pywt.cwt(sig, escalas, 'cmor3-3', sampling_period = (1 / fs))
```

```python
# dado que cmor3-3 genera numeros complejos, calcular el modulo
x = abs(coef)
```

```python
plt.pcolormesh(time, freqs, x)
plt.xlabel('Time [sec]')
plt.ylabel('Frequency [Hz]')
plt.show()
```

```python
print(x.shape)
plt.matshow(x)
plt.show()
```
