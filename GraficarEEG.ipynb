```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#la muestra 2 es mano derecha para el train del sujeto 1
eeg = pd.read_csv("MI-EEG-B1T.csv", header=None).iloc[1, :]
```

```python
c3 = eeg[0:1000]
cz = eeg[1000:2000]
c4 = eeg[2000:3000]
```

```python
# 4 segundos con fs=250Hz
fs = 250
time = np.arange(0, 4, 1/fs)

plt.plot(time, c3)
plt.plot(time, cz)
plt.plot(time, c4)
plt.grid()
plt.xlabel('Tiempo')
plt.ylabel('Amplitud')
plt.legend(['C3', 'Cz', 'C4'])
plt.show()
```
