#### Escalograma con CWT de shirp

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.signal import chirp
import pywt
```

```python
fs = 250
time = np.arange(0, 10, 1/fs)
sig = chirp(time, f0=6, f1=1, t1=10)
sig.shape
```

```python
plt.plot(time, sig)
plt.grid()
plt.xlabel('Time')
plt.ylabel('Amplitude')
plt.show()
```

### Morlet

```python
# calcular las escalas de 1-10Hz para Morlet wavelets
# f = scale2frequency(wavelet, scale)/sampling_period
escalas = pywt.scale2frequency('morl', np.arange(1,11)) / (1/fs)
print(escalas)
```

```python
# Escalograma (CWT)
# pywt.cwt(senal, escalas, tipo wavelet, periodo de muestreo = (1 / fs))
coef, freqs = pywt.cwt(sig, escalas, 'morl', sampling_period = (1 / fs))
```

```python
plt.pcolormesh(time, freqs, coef)
plt.xlabel('Time [sec]')
plt.ylabel('Frequency [Hz]')
plt.show()
```

```python
print(coef.shape)
plt.matshow(coef)
plt.show()
```

### Complex Morlet

```python
# calcular las escalas de 1-10Hz para complex Morlet wavelets
# f = scale2frequency(wavelet, scale)/sampling_period
escalas = pywt.scale2frequency('cmor3-3', np.arange(1,11)) / (1/fs)
print(escalas)
```

```python
# Escalograma (CWT)
# pywt.cwt(senal, escalas, tipo wavelet, periodo de muestreo = (1 / fs))
coef, freqs = pywt.cwt(sig, escalas, 'cmor3-3', sampling_period = (1 / fs))
```

```python
# dado que cmor3-3 genera numeros complejos, calcular el modulo
x = abs(coef)
```

```python
plt.pcolormesh(time, freqs, x)
plt.xlabel('Time [sec]')
plt.ylabel('Frequency [Hz]')
plt.show()
```

```python
print(x.shape)
plt.matshow(x)
plt.show()
```
