#### Espectrograma sin y con STFT de seno

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.signal import spectrogram
```

```python
# 8 segundos de una señal con fs=250Hz
# la señal es la suma de tres señales senos

fs = 250
time = np.arange(0, 8, 1/fs)
x1 = np.sin(2*np.pi*4*time) #seno a 4Hz
x2 = np.sin(2*np.pi*16*time) #seno a 16Hz
x3 = np.sin(2*np.pi*32*time) #seno a 32Hz
sig = x1 + x2 + x3
```

```python
plt.plot(time, sig)
plt.grid()
plt.xlabel('Time')
plt.ylabel('Amplitude')
plt.show()
```

```python
powerSpectrum, freqenciesFound, t, imageAxis = plt.specgram(sig, Fs=fs)
plt.xlabel('Time [sec]')
plt.ylabel('Frequency [Hz]')
plt.show()
```

```python
plt.pcolormesh(t, freqenciesFound, powerSpectrum)
plt.xlabel('Time [sec]')
plt.ylabel('Frequency [Hz]')
plt.show()
```

```python
f, t, Sxx = spectrogram(sig, fs, window='hann', nperseg=250, noverlap=225)
plt.pcolormesh(t, f, Sxx)
plt.xlabel('Time [sec]')
plt.ylabel('Frequency [Hz]')
plt.show()
```
